```python
print(df.columns)
print(df.dtypes)
```

```python
null_values = df.isnull().sum()
print("Null values in each column:\n", null_values)
```

```python
total_null = df.isnull().sum().sum()
print("\nTotal number of null values:", total_null)
```

```python
# Count the number of duplicate rows
duplicate_values = df.duplicated().sum()
print("\nNumber of duplicate rows:", duplicate_values)
```

```python
# Display the duplicate rows, if needed
duplicates = df[df.duplicated()]
print("\nDuplicate rows:\n", duplicates)
```

```python
# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Verify the number of rows after removing duplicates
print("\nNumber of rows after removing duplicates:", len(df))
```

```python
# Save the modified DataFrame to a new file
df.to_csv("cleaned_file.csv", index=False)
```

```python
import google.colab.files as files
import pandas as pd
uploaded = files.upload()
df = pd.read_csv('remove_col.csv')
month_mapping={'Jan':1,'Feb':2,'Mar':3,'Apr':4,'May':5,'June':6,'Jul':7,'Aug':8,'Sep':9,'Oct':10,'Nov':11,'Dec':12}
df['Month']=df['Month'].map(month_mapping)
visitor_type_mapping={'New_Visitor':1,'Returning_Visitor':2,'Other':3}
df['VisitorType']=df['VisitorType'].map(visitor_type_mapping)
df['Revenue'] = df['Revenue'].astype(int)
df.to_csv('numeric_col.csv', index=False)
files.download('numeric_col.csv')
```
